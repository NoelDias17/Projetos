```python
import pandas as pd

import yagmail

import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
from tkcalendar import DateEntry

import warnings

warnings.filterwarnings('ignore')

#titulo e tamanho das janelas
data_vence_amanha = tk.Tk()

data_vence_amanha.title('VENCE AMANHÃ')

largura = 240
altura = 280

data_vence_amanha.geometry("%dx%d+%d+%d" % (largura,altura,500,250))

#data relatorios
label_escolher_data_inicial = tk.Label(anchor='n',text='Favor insira a data INICIAL do relatório:', borderwidth=5, bg='#CCCCCC')
label_escolher_data_inicial.grid(row=0, column=0, columnspan=1, padx=10, pady=10, sticky='NSEW')

data_relatorio_inicial = DateEntry(year=2022, locale='pt_br')
data_relatorio_inicial.grid(row=1, column=0, columnspan=1, padx=10, pady=10, sticky='W')

label_escolher_data_final = tk.Label(anchor='n',text='Favor insira a data FINAL do relatório:', borderwidth=5, bg='#CCCCCC')
label_escolher_data_final.grid(row=2, column=0, columnspan=1, padx=10, pady=10, sticky='NSEW')

data_relatorio_final = DateEntry(year=2022, locale='pt_br')
data_relatorio_final.grid(row=3, column=0, columnspan=1, padx=10, pady=10, sticky='W')

def venc_gsc():
    vencimentos = pd.read_excel(r"S:\CLINICA\Noel\Controle Vencimentos.xlsm")
    data_inical = data_relatorio_inicial.get() 
    data_final = data_relatorio_final.get()
    
    novo_header = vencimentos.iloc[4]
    novo_header = list(novo_header)
    vencimentos.columns = novo_header
    vencimentos = vencimentos.drop([0,1,2,3,4])

    vencimentos['empresa'] = vencimentos['empresa'].str.strip()
    
    vencimentos = vencimentos[(vencimentos['agendado para'] == data_inical) |
                              (vencimentos['agendado para'] == data_final) ]
    
    vencimentos = vencimentos[vencimentos['empresa'] == 'GSC']
    
    vencimentos = vencimentos[['fornecedor','valor']]
    
    vencimentos.columns = ['GSC','-']
    
    vencimentos = vencimentos.set_index('GSC')
    
    display(vencimentos)
    
    # Enviando relatório por email
    usuario = yagmail.SMTP('#email','#senha')

    datatextovenceamanha = data_relatorio_final.get()[0:2]+data_relatorio_final.get()[2:5]
    textovenceamanha = 'Vence Amanhã {}'.format(datatextovenceamanha)
    conteudoemail = [textovenceamanha, vencimentos]

    usuario.send(to='noeldias17@gmail.com', subject='Vence Amanhã GSC', contents=conteudoemail)
    
    print('Vencimentos GSC enviados com sucesso!')

def venc_cms():
    vencimentos = pd.read_excel(r"S:\CLINICA\Noel\Controle Vencimentos.xlsm")
    data_inical = data_relatorio_inicial.get() 
    data_final = data_relatorio_final.get()
    
    novo_header = vencimentos.iloc[4]
    novo_header = list(novo_header)
    vencimentos.columns = novo_header
    vencimentos = vencimentos.drop([0,1,2,3,4])
    
    vencimentos['empresa'] = vencimentos['empresa'].str.strip()
    
    vencimentos = vencimentos[(vencimentos['agendado para'] == data_inical) |
                              (vencimentos['agendado para'] == data_final) ]
    
    vencimentos = vencimentos[vencimentos['empresa'] == 'CMS']
    
    vencimentos = vencimentos[['fornecedor','valor']]
    
    vencimentos.columns = ['CMS','-']
    
    vencimentos = vencimentos.set_index('CMS')
    
    display(vencimentos)
    
    # Enviando relatório por email
    usuario = yagmail.SMTP('#email','#senha')

    usuario.send(to='noeldias17@gmail.com', subject='Vence Amanhã CMS', contents=vencimentos)
    
    print('Vencimentos CMS enviados com sucesso!')

botao_gerar_gsc = tk.Button(text='Gerar Relatórios GSC', command=venc_gsc, height=1, width=20)
botao_gerar_gsc.grid(row=4, column=0, columnspan=1, padx=10, pady=10)

botao_gerar_cms = tk.Button(text='Gerar Relatórios CMS', command=venc_cms, height=1, width=20)
botao_gerar_cms.grid(row=5, column=0, columnspan=1, padx=10, pady=10)

data_vence_amanha.mainloop()
```
