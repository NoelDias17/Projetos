### GSC

```python
import pandas as pd
import yagmail
import time
import pyautogui

from pathlib import Path
from os.path import getmtime

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains

# Descobrir valor da nota
# Verificar se nota ja foi emitida anteriormente (asa e nota carioca)
# preparar planilha com dados do paciente

# abrir navegador e logar no site da nota carioca
servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)

navegador.get(r'https://notacarioca.rio.gov.br/senhaweb/login.aspx?ReturnUrl=%2fcontribuinte%2fConsulta.aspx')
navegador.maximize_window()
navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbCpfCnpj').send_keys('#LoginNotaCarioca')
navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbSenha').send_keys('#SenhaNotaCarioca')
pyautogui.alert('Por Favor, preencha o captcha e logue no site!')

# importar planilha e pegar informações (precisei importar depois de abrir o navegador para funcionar)
nf_paciente = pd.read_excel(r'S:\CLINICA\Noel\NFs Pacientes.xlsx')

nome = nf_paciente.loc[0,'nome']
cpf = nf_paciente.loc[0,'cpf']
valor = nf_paciente.loc[0,'valor']
email = nf_paciente.loc[0,'email']
servico = nf_paciente.loc[0,'servico']

display(servico)

navegador.find_element(By.ID,'menuDivAux').click()
navegador.find_element(By.XPATH,'//*[@id="ctl00_cphCabMenu_pnEmite"]/p/a[1]').click()

# CPF Precisa estar com pontos e traços
navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbCPFCNPJTomador').send_keys(cpf)
navegador.find_element(By.NAME,'ctl00$cphCabMenu$btAvancar').click()
time.sleep(1)
navegador.execute_script("window.scrollTo(0,600)")
time.sleep(1)

# lógica para preenchimento do texto da nota
if servico == 'vacina':
    navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbDiscriminacao').send_keys('Formulação de extrato alergênico para imunoterapia (uso em consultório)')
elif servico == 'consulta':
    navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbDiscriminacao').send_keys('Consulta médica em ambulatório prestado ao próprio.')
elif servico == 'histopatologico':
    navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbDiscriminacao').send_keys('Análise de histopatológico prestada ao próprio.')
elif servico == 'teste':
    navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbDiscriminacao').send_keys('Teste alérgico ambulatorial prestado ao próprio.')
else:
    navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbDiscriminacao').send_keys('Pequena cirurgia ambulatorial prestado ao próprio.')

navegador.find_element(By.NAME,'ctl00$cphCabMenu$tbValor').send_keys(valor)

# Emitir nota
navegador.find_element(By.NAME,'ctl00$cphCabMenu$btEmitir').click()

# Aceitar alerta que confirma emissão
alerta = navegador.switch_to.alert
alerta.accept()

# Verificar Nota
pyautogui.alert('Tudo Certo com a nota ?')

# Fazer Download do PDF
time.sleep(3)
navegador.find_element(By.NAME,'ctl00$cphBase$btGerarPDF').click()
time.sleep(3)

# definir caminho da pasta downloads
nota_fiscal = Path(r'C:\Users\Administrator\Downloads')

# definir formato que será pesquisado
nota_fiscal = nota_fiscal.glob('*.pdf')

# definir o arquivo mais recente a entrar na pasta downloads
arquivo_mais_recente = max(nota_fiscal, key=getmtime)

# Definir texto do email
conteudo_email = 'Boa tarde sr(a) {}\n\nSegue em anexo a Nota fiscal Solicitada\n\nQualquer dúvida estamos a disposição\n\nAtt Noel'.format(nome)

# enviar por email
usuario = yagmail.SMTP('#email','#senha')
usuario.send(to=email, subject='Nota Fiscal Solicitada', contents=conteudo_email, attachments=arquivo_mais_recente)

print('Email enviado com sucesso!')

# Fechar navegador
navegador.quit()
```
